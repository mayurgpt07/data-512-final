### Converting non numerical variables

To identify correlation between variables we need to first convert all objects into numerical format. We do this in a two step format:
- Convert object type variables into category type variables
- Use a **LabelEncoder** from the scikit-learn pacakge to assign a numerical value corresponding to each category

The **LabelEncoder.fit(dataframe_name.column_name)** checks for the different values in a column and **LabelEncoder.transform(dataframe_name.column_name)** assigns a numerical value corresponding to them

```python
##Converting objects labels into categorical
insurance_df[['sex', 'smoker', 'region']] = insurance_df[['sex', 'smoker', 'region']].astype('category')
```

```python
##Converting category labels into numerical using LabelEncoder
label = LabelEncoder()
label.fit(insurance_df.sex.drop_duplicates())
insurance_df.sex = label.transform(insurance_df.sex)
label.fit(insurance_df.smoker.drop_duplicates())
insurance_df.smoker = label.transform(insurance_df.smoker)
label.fit(insurance_df.region.drop_duplicates())
insurance_df.region = label.transform(insurance_df.region)
insurance_df.dtypes
```

### Feature Importance Matrix
Random Forest Regressor have an ability to identify feature importance on the data they are used. We will use this feature to verify
- ***smoker*** is an important variable or not
- Are there any other important variables

### Standardize the variables
Before using any model, it is important to standardize the values. The standardizing can be given as:
\begin{equation*}
x =
\frac{x - \mu(x)} {\sigma(x)}
\end{equation*}

```python
## Create a copy of dataframe
df_Scaled = insurance_df.copy()
```

```python
## Standardize all numerical columns
col_names = ['age', 'bmi', 'children', 'charges']
features = df_Scaled[col_names]
scaler = StandardScaler().fit(features.values)
features = scaler.transform(features.values)

df_Scaled[col_names] = features
print(df_Scaled)
```

```python
## Split the data into training an testing data
x = df_Scaled.drop(['charges'], axis=1)
y = df_Scaled.charges
x_train, x_test, y_train, y_test = holdout(x, y, test_size=0.2, random_state=0)
```

### Machine Learning models

The following machine learning models are used to model data in the analysis:
- Linear Regression - Models relations by fitting linear equation on linear variables
- Random Forest Regression - Models relations by fitting tree based model
- Polynomial Regression - Models relation by fitting linear equation on non linear variables
- L1-regularized Polynomial Regression - Models relation by fitting penalized linear equation on non linear variables

For all machine learning model the **Model.fit(set_of_independent_features, dependent_output)** method trains the algorithm on the dataset, while **Model.predict(set_of_independent_features)** predicts the future output

<font size=4><p style="color:purple"> Random Forest Regressor

```python
## Fit Random Forest model and determine MSE
Rfr = rfr(n_estimators = 100, criterion = 'mse',
                              random_state = 1,
                              n_jobs = -1)
Rfr.fit(x_train,y_train)
x_train_pred = Rfr.predict(x_train)
x_test_pred = Rfr.predict(x_test)

print('MSE train data: %.3f, MSE test data: %.3f' % 
      (metrics.mean_squared_error(x_train_pred, y_train),
       metrics.mean_squared_error(x_test_pred, y_test)))
```

```python
print('Feature importance ranking\n\n')
importances = Rfr.feature_importances_
std = np.std([tree.feature_importances_ for tree in Rfr.estimators_],axis=0)
indices = np.argsort(importances)[::-1]
variables = ['age', 'sex', 'bmi', 'children','smoker', 'region']
importance_list = []
for f in range(x.shape[1]):
    variable = variables[indices[f]]
    importance_list.append(variable)
    print("%d.%s(%f)" % (f + 1, variable, importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(importance_list, importances[indices],
       color="y", yerr=std[indices], align="center")
```

**Insights:** Based on the plot of feature importance, we can infer that **smoker** is the most important variable for medical insurance cost. The plot also brings ***bmi*** and ***age*** as two new variables which impact charges

### Permutation Feature importance

***Permutation feature importance*** is a model inspection technique that can be used for any fitted estimator
This procedure breaks the relationship between the feature and the target, thus the drop in the model score is indicative of how much the model depends on the feature.
<br />
<br />

#### Our Experiment:

We use the Random Forest as the estimator for permutation feature importance. The selection of the model was based on the fact that Random Forest can capture both linear and non-linear dependencies of features. Therefore it is a more exhaustive model


```python
perm = PermutationImportance(Rfr, random_state=1).fit(x_test, y_test)
eli5.show_weights(perm, feature_names = x_test.columns.tolist())
```

**Insights:** Permutation testing reconfirms that being a smoker is the biggest reason for higher medical insurance cost, followed by bmi and age.

### Research Question - 2
<font size=4 > <p style="color:purple"> Can we create a model that can compute the medical insurance cost of individual? </p></font>

We implement different models and compare them to identify which fits our data best. Since charges are a continuous variable we use Mean Squared Error (MSE) ro compare the different models.

To put in simple words, MSE is the average squared difference between the estimated values and the actual value.

We fit each model and then check the test data MSE and the model with the minimum test data MSE will be our best estimator.

##### Note: Random Forest implemented above is also considered in this

<font size=4><p style="color:purple"> Linear Regression

```python
## Fit Linear Regression model and determine MSE
x = df_Scaled.drop(['charges'], axis = 1)
y = df_Scaled['charges']
Lin_reg = LinearRegression()
Lin_reg.fit(x_train, y_train)
print('The intercept value: ',Lin_reg.intercept_)
print('The coefficients value: ',Lin_reg.coef_)
print('The fit score based on training data: ',Lin_reg.score(x_test, y_test))
```

```python
x_train_pred = Lin_reg.predict(x_train)
x_test_pred = Lin_reg.predict(x_test)
print('MSE train data: %.3f, MSE test data: %.3f' % 
      (metrics.mean_squared_error(x_train_pred, y_train),
       metrics.mean_squared_error(x_test_pred, y_test)))
```

The result we got is good enough, but we can try to improve it a bit by reducing unimportant features later

<font size=4><p style="color:purple"> L1-Regularized Linear Regression

```python
## Fit Linear Regression model and determine MSE
x = df_Scaled.drop(['charges'], axis = 1)
y = df_Scaled['charges']
Lasso_reg = Lasso()
Lasso_reg.fit(x_train, y_train)
print('The intercept value: ',Lin_reg.intercept_)
print('The coefficients value: ',Lin_reg.coef_)
print('The fit score based on training data: ',Lin_reg.score(x_test, y_test))
```

```python
x_train_pred = Lasso_reg.predict(x_train)
x_test_pred = Lasso_reg.predict(x_test)
print('MSE train data: %.3f, MSE test data: %.3f' % 
      (metrics.mean_squared_error(x_train_pred, y_train),
       metrics.mean_squared_error(x_test_pred, y_test)))
```

<font size=4><p style="color:purple"> Polynomial Regression

```python
## Fit Polynomial Regression model and determine MSE
x = df_Scaled.drop(['charges', 'sex', 'region'], axis = 1)
y = df_Scaled.charges
pol = PolynomialFeatures (degree = 2)
x_pol = pol.fit_transform(x)
x_train, x_test, y_train, y_test = holdout(x_pol, y, test_size=0.2, random_state=0)
Pol_reg = LinearRegression()
Pol_reg.fit(x_train, y_train)
y_train_pred = Pol_reg.predict(x_train)
y_test_pred = Pol_reg.predict(x_test)
print('The intercept value: ',Pol_reg.intercept_)
print('The coefficients value: ',Pol_reg.coef_)
print('The fit score based on training data: ',Pol_reg.score(x_test, y_test))
```

```python
x_train_pred = Pol_reg.predict(x_train)
x_test_pred = Pol_reg.predict(x_test)
print('MSE train data: %.3f, MSE test data: %.3f' % 
      (metrics.mean_squared_error(x_train_pred, y_train),
       metrics.mean_squared_error(x_test_pred, y_test)))
```

<font size=4><p style="color:purple"> L1-Regularized Polynomial Regression

```python
## Fit L1 regularized polynomial regression model and determine MSE
Lasso_poly_reg = Lasso()
Lasso_poly_reg.fit(x_train, y_train)
print('The intercept value: ',Lasso_poly_reg.intercept_)
print('The coefficients value: ',Lasso_poly_reg.coef_)
print('The fit score based on training data: ',Lasso_poly_reg.score(x_test, y_test))
```

```python
x_train_pred = Lasso_poly_reg.predict(x_train)
x_test_pred = Lasso_poly_reg.predict(x_test)
print('MSE train data: %.3f, MSE test data: %.3f' % 
      (metrics.mean_squared_error(x_train_pred, y_train),
       metrics.mean_squared_error(x_test_pred, y_test)))
```
